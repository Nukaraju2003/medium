```python
def openapicall(prompt):
  model_engine = "gpt-3.5-turbo"
  response = openai.ChatCompletion.create(
      model= model_engine,
      temperature=0.5,
      messages=[
          {"role": "system", "content": "You are a helpful assistant."},
          {"role": "user", "content": prompt},
      ])
  return response
```

```python
#content = "Relevance: 80/100 - The answer provides relevant information about the candidate's experience in the IT industry and their exposure to different domains. However, the answer could be more focused on the question of what inspired them to pursue a career in technology.\n\nFluency: 70/100 - The answer is somewhat disjointed and lacks a clear structure. The candidate's thoughts seem to jump around a bit, and there are several instances where they pause or stumble over their words.\n\nClarity: 60/100 - The answer is not very clear. The candidate's thoughts are not well-organized, and they use a lot of filler words and phrases. As a result, it's difficult to discern a clear answer to the question.\n\nConfidence: 80/100 - The candidate seems confident in their abilities and experience, but this confidence is somewhat undermined by the lack of clarity and fluency in their answer.\n\nEnthusiasm: 70/100 - The candidate seems somewhat enthusiastic about their career in technology, but this enthusiasm is not very well conveyed due to the lack of clarity and structure in their answer.\n\nUnique Qualities: 70/100 - The candidate mentions their experience in people management as well as their approachability and quick learning abilities, which are unique qualities that could set them apart from other candidates. However, these qualities are not emphasized very strongly in the answer.\n\nProfessionalism: 80/100 - The candidate's answer is generally professional in tone, although the lack of clarity and structure could be seen as somewhat unprofessional.\n\nExplanation: Overall, the candidate's answer provides some relevant information about their experience in the IT industry and their exposure to different domains. However, the answer lacks clarity and structure, which makes it difficult to discern a clear answer to the question of what inspired them to pursue a career in technology. The candidate's enthusiasm and unique qualities are somewhat undermined by the lack of clarity and fluency in their answer. Nevertheless, the answer is generally professional in tone, and the candidate seems confident in their abilities and experience."
def parsable(content):
  pub_set = { 'Relevance', 'Fluency', 'Clarity', 'Confidence', 'Enthusiasm', 'Unique qualities', 'Professionalism', 'Overall'}
  json_data =dict()
  lines = content.split('\n\n')
  for line in lines:
    if "Explanation" in line:
      continue
    parse = line.strip().split(':')
    for key_ in pub_set:
      if key_ in parse[0]:
        key = key_

    print(parse)
    score = int(parse[1][:7].split('/')[0])
    s = parse[1][7:]
    for i in range(len(s)):
      if 'A' <= s[i] <= 'Z' or 'a'<= s[i] <='z':
        break
    explin = s[i:]
    json_data[key] = (score, explin)
  return json_data
```
