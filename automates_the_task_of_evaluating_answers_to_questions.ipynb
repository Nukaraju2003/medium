```python
def openapicall(prompt):
  model_engine = "gpt-3.5-turbo"
  response = openai.ChatCompletion.create(
      model= model_engine,
      temperature=0.5,
      messages=[
          {"role": "system", "content": "You are a helpful assistant."},
          {"role": "user", "content": prompt},
      ])
  return response
```

```python
def parsable(content):
  pub_set = { 'Relevance', 'Fluency', 'Clarity', 'Confidence', 'Enthusiasm', 'Unique qualities', 'Professionalism', 'Overall'}
  json_data =dict()
  lines = content.split('\n\n')
  for line in lines:
    if "Explanation" in line:
      continue
    parse = line.strip().split(':')
    for key_ in pub_set:
      if key_ in parse[0]:
        key = key_

    print(parse)
    score = int(parse[1][:7].split('/')[0])
    s = parse[1][7:]
    for i in range(len(s)):
      if 'A' <= s[i] <= 'Z' or 'a'<= s[i] <='z':
        break
    explin = s[i:]
    json_data[key] = (score, explin)
  return json_data
```

```python
import csv
import openai  
openai.api_key = "Your API key"

# open the CSV file to read data
with open('/content/drive/MyDrive/Dr. Abhinav Dayal Sir tasks/evaluation task/questionandanswers.csv') as csv_file:
    # create a CSV reader object
    csv_reader = csv.reader(csv_file)
    # skip the header row
    next(csv_reader)
    
    # open the CSV file to write output data
    with open('/content/score.csv', mode='w') as output_file:

        # create a CSV writer object
        fieldnames = ['Relevance Score', 'Relevance explanation', 'Fluency Score', 'Fluency explanation','Clarity Score','Clarity explanation', 'Confidence Score', 'Confidence explanation', 'Enthusiasm Score', 'Enthusiasm explanation','Unique qualities Score','Unique qualities explanation', 'Professionalism Score', 'Professionalism explanation', 'Overall Score', 'Overall explanation']
        csv_writer = csv.DictWriter(output_file, fieldnames=fieldnames)
        # write the header row to the output file
        csv_writer.writeheader()
        
        # loop over the rows in the CSV file
        for i, row in enumerate(csv_reader, 1):
            # extract the values from the row
            column1_value = row[0]
            column2_value = row[1]      
            # create prompt using column1_value as the question and column2_value as the answer
            prompt = f"For output_filethe question, \"{column1_value}\", give a Score out of 100 each for Relevance, fluency, Clarity, Confidence, Enthusiasm, Unique qualities,Professionalism, Overall for the following answer with a proper explanation for each: \"{column2_value}\""
            response = openapicall(prompt)
            message = response.choices[0]['message']['content']
            print(message.strip())
            json_data = parsable(message.strip())
            # print("{}: {}".format(message['role'], message['content']))
            
             # extract the scores and explanations from the AI response and write them to the output file
            print('message check', message)
            # user_scores = message.strip()
            # user_scores_list = user_scores.split(',')
            user_scores_dict = {}
            for key, values in json_data.items():
              user_scores_dict[str(key)+' Score'] = values[0]
              user_scores_dict[str(key)+' explanation'] = values[1]

            # write the user scores and explanations to the output file
            print(user_scores_dict)
            csv_writer.writerow(user_scores_dict)
```
